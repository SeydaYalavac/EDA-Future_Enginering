# Understanding Missing Values in Feature Engineering

When working with real-world data, missing values are a common problem. In feature engineering, handling missing values correctly is crucial because they can affect the performance of machine learning models.

## Types of Missing Values
Missing values can be categorized into three main types:

1. **Missing Completely at Random (MCAR):** The missing values occur randomly and have no relationship with any other variable.
   * *Example:* A person forgets to fill in their age in a survey.
   * *Effect:* No bias is introduced; simple imputation (mean/median) is usually fine.
2. **Missing at Random (MAR):** Related to other observed variables but not the missing data itself.
   * *Example:* Men being less likely to report weight in a medical dataset.
3. **Missing Not at Random (MNAR):** The reason for missingness is related to the missing value itself.
   * *Example:* High earners not reporting their salaries.

---

## Summary Table

| Type | Example | Best Handling Method |
| :--- | :--- | :--- |
| **MCAR** | Randomly missing age values | Drop rows or simple imputation |
| **MAR** | Income missing based on education | Regression or KNN Imputation |
| **MNAR** | High earners hiding salaries | Domain knowledge or modeling missingness |

## Handling Missing Values: Practical Applications

### 1. Removing Missing Values
If the missing data is minimal, we can drop them.

```python
# SatÄ±r silme denemesi (TÃ¼m NaN olan satÄ±rlar gider)
print(f"SatÄ±r silindikten sonra kalan veri: {df.dropna().shape}")

# SÃ¼tun silme (Ä°Ã§inde NaN olan tÃ¼m sÃ¼tunlarÄ± atar)
df_dropped_cols = df.dropna(axis=1)
df_dropped_cols.head()
```

### 2. Imputation (Filling Missing Values)

* **Mean Imputation:** Best for normally distributed data.
* **Median Imputation:** Best when outliers are present.
* **Mode Imputation:** Used for categorical data (like `embarked`).

```python
# SayÄ±sal DeÄŸerler Ä°Ã§in GÃ¶rselleÅŸtirme
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()

# Mean ve Median UygulamasÄ±
df['Age_mean'] = df['age'].fillna(df['age'].mean())
df['age_median'] = df['age'].fillna(df['age'].median())

# Kategorik DeÄŸerler Ä°Ã§in Mode (En sÄ±k tekrar eden) UygulamasÄ±
mode_value = df['embarked'].mode()[0]
df['embarked_mode'] = df['embarked'].fillna(mode_value)

# SonuÃ§larÄ±n KontrolÃ¼
print(f"Embarked sÃ¼tununda kalan eksik deÄŸer: {df['embarked_mode'].isnull().sum()}")
df[['age', 'Age_mean', 'age_median', 'embarked', 'embarked_mode']].head(10)
```
